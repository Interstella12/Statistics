# 思考题的个人解答

1.预处理包括哪些内容?

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;数据审核，筛选，排序，透视图
</font>

2.分类数据和顺序数据的整理和图示方法各有哪些?

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;(1) 分类数据的整理有: 频数统计和数据透视表；分类数据的图示方法有: 条形图(bar chart)，帕累托图(Pareto chart)，饼图(Pie chart)，　环形图(Doughnut chart，适用于多样本比较)．
<br/>
&ensp;&ensp;&ensp;&ensp;(2) 顺序数据的整理有: 频数统计，累计频数/频率及数据透视表；图示方法有:  条形图(bar chart)，　帕累托图(Pareto chart)，饼图(Pie chart)，　环形图(Doughnut chart，适用于多样本比较)及累积频数分布或者频率图．
</font>

3.数值型数据点的分组方法有哪些?简述组距分组的步骤．

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;分组方法有单变量值分组和组距分组．其中组距分组步骤是: (1)确定组数，建议5-15组之间；(2)确定各组组距，一般是全体数据的最大值减去最小值再除以组数确定，如果出现极端的最大值或者最小值，应该灵活选择上下限；(3)编制分组后的频数分布表
</font>

4.直方图和条形图有何区别？

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;(1)条形图的宽度没有含义，直方图的含义代表了组距;(2)直方图的各矩形是连续排列的，而条形图是分开排列;(3)条形图用于展示分类数据，而直方图用于展示数值型数据
</font>

5.绘制线图应注意哪些问题？

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;一般把时间绘在横轴，观测值绘在纵轴，长宽比采用10:7(此比例应该是考虑标题，刻度的原因，因此大于黄金比例)，一般情况下下，纵轴下端应从0开始，如果数据与0间距过大，可以采用折断符号将纵轴折断
</font>

6.饼图和环形图有什么区别？

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;环形图中间多一个空洞，每个样本用一个环来表示，每个环的一段表示一部分数据，因此除了样本内可以清楚表示出比例分布，还能直观得看到样本间的分布比较
</font>

7.茎叶图和直方图相比有什么优点？他们的应用场合是什么？

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;茎叶图保留了原始数值的展示，适用于小批量数据，直方图适用于大批量数据
</font>

8.鉴别图标优劣的准则有哪些？

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;(1)精心设计；(2)复杂的观点得到简明，确切，高效的阐述；(3)读者在最短的时间获得最大量的信息；(4)多维；(5)表述数据的真实情况
</font>

9.制作统计表时应注意哪几个问题？

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;(1)合理安排结构；(2)表头赢包括表号，总标题和表中数据的单位；(3)表格上下两条线加粗，两边不封口；(4)必要时增加注释
</font>

### 指出上面数据属于什么类型?

<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;品质数据
</font>

### 用Excel制作一张频数分布表


```python
#
data = pd.read_excel('../source/《统计学》（第7版）—例题习题数据/（03）第3章  数据的图表展示（7）/（3-2）本章练习题数据/习题3.1.xlsx')
data.reset_index(inplace=True)
count = data.groupby('评价等级').count()
count.columns = ['频数']
count.reset_index(inplace=True)
count['百分比'] = count['频数'].apply(lambda x: x/count['频数'].sum())
count['累积百分比'] = count['百分比'].cumsum()
count
```

### 绘制一个条形图，反映评价等级的分布

```python
#
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
_ = sns.countplot(data=data[['评价等级']],x='评价等级',order=sorted(data['评价等级'].unique().tolist()))
```

### 绘制评价等级的帕累托图

```python
#

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
sns.countplot(data=data[['评价等级']],x='评价等级',order=sorted(data['评价等级'].unique().tolist()))
yticks = np.linspace(0, ax.get_ylim()[1], 6, endpoint=True)
yticks_labels = [f'{ytick:.0f}' for ytick in yticks]
ax.set_yticks(yticks)
ax.set_yticklabels(yticks_labels)

ax_2 = ax.twinx()
sns.lineplot(count['评价等级'], count['累积百分比'], markers='o', markersize=10,)
yticks = np.linspace(0, 1, 6, endpoint=True)
yticks_labels = [f'{ytick*100:.0f}%' for ytick in yticks]
ax_2.set_ylim((0, 1))
ax_2.set_yticks(yticks)
_ = ax_2.set_yticklabels(yticks_labels)
```

## 3.2
为了确定灯泡的使用寿命，在一批灯泡中随机抽取100个进行测试，所得结果如下:
  ![灯泡寿命](../source/pics/chapter3/习题3.2.png '灯泡寿命')

### 利用计算机对上面的数据进行排序

```python
data = pd.read_excel('../source/《统计学》（第7版）—例题习题数据/（03）第3章  数据的图表展示（7）/（3-2）本章练习题数据/习题3.2.xlsx')
print(sorted(data['使用寿命'].tolist()))
```

### 以组距为10进行等距分组，整理成频数分布表

```python
# 
data['分组'] = pd.cut(data['使用寿命'], bins=np.linspace(start=650,stop=750, num=11))
count = data.groupby('分组').count()
count.reset_index(inplace=True)
count.columns = ['分组', '频数']
count['百分比'] = count['频数'].apply(lambda x: x/count['频数'].sum())
count['累积百分比'] = count['百分比'].cumsum()
count
```

### 根据分组数据绘制直方图，说明数据分布的特点
<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;数据呈右偏态
</font>

```python
# 
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
bins = np.linspace(start=650,stop=750, num=11)
sns.distplot(data['使用寿命'], hist=True, bins=bins, kde=False)
ax.set_xticks((bins+5)[:-1])
xtickslabels = [f'[{bins[i]:.0f}, {bins[i+1]:.0f})' for i in range(10)]
_ = ax.set_xticklabels(xtickslabels, rotation=45)
```

### 制作茎叶图，并与直方图做比较

```python
# 
data['stem'] = data['使用寿命'].apply(lambda x: x//10)
data['leaf'] = data['使用寿命'].apply(lambda x: x%10)
gp = data.groupby('stem')
print('灯泡使用寿命 Stem-and-leaf Plot\n')
print('{:10s} {:4s} | {:100s}'.format('Frequency', 'stem', 'leaf'))
for stem, df in gp:
    leaves = sorted(df["leaf"].tolist())
    frequency = len(leaves)
    leaves = [str(leaf) for leaf in leaves]
    leaves = "".join(leaves)
    print(f'{frequency:10d} {stem:4d} | {leaves}')
```

## 3.5
甲，乙两个班各有40名学生，期末统计学考试成绩的分布如下: 
![统计学成绩](../source/pics/chapter3/习题3.5.png '统计学成绩')

### 根据上面数据，画出两个班考试成绩的对比条形图和环形图

```python
#
data = pd.read_excel('../source/《统计学》（第7版）—例题习题数据/（03）第3章  数据的图表展示（7）/（3-2）本章练习题数据/习题3.5.xlsx', skiprows=1)
data.columns = ['成绩', '甲班', '乙班']
data_jia = data[['成绩', '甲班']].copy()
data_jia['班级'] = '甲班'
data_jia.columns = ['成绩', '人数', '班级']
data_jia['比例'] = data_jia['人数'].apply(lambda x: x/ data_jia['人数'].sum())
data_yi = data[['成绩', '乙班']].copy()
data_yi['班级'] = '乙班'
data_yi.columns = ['成绩', '人数', '班级']
data_yi['比例'] = data_yi['人数'].apply(lambda x: x/ data_yi['人数'].sum())
data = pd.concat([data_jia, data_yi])
data.index = range(data.shape[0])

# 条形图
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot(121)
_ = sns.barplot(data=data, x='成绩', y='人数', hue='班级')

# 环形图

ax = fig.add_subplot(122)
colors = ['c', 'r', 'y', 'g', 'gray']
# 外环
wedges1, texts1, autotexts1 = plt.pie(data_jia['比例'],
    autopct = '%3.1f%%',
    radius = 1,
    pctdistance = 0.85,
    startangle = 180,
    colors = colors,
    textprops = {'color': 'w'},
    wedgeprops = {'width': 0.3, 'edgecolor': 'w'}
)

# 内环
wedges2, texts2, autotexts2 = plt.pie(data_yi['比例'],
    autopct = '%3.1f%%',
    radius = 0.7,
    pctdistance = 0.75,
    startangle = 180,
    colors = colors,
    textprops = {'color': 'w'},
    wedgeprops = {'width': 0.3, 'edgecolor': 'w'}
)

ax.set_xlabel('外环甲班，内环乙班')
# 图例
plt.legend(wedges1,
          data_jia['成绩'],
          fontsize = 12,
          title = '成绩',
          loc = 'center right',
          bbox_to_anchor = (1.0, 0.9))

# 设置文本样式
plt.setp(autotexts1, size=15, weight='bold')
plt.setp(autotexts2, size=15, weight='bold')
plt.setp(texts1, size=15)

# 标题
plt.suptitle('统计学成绩', fontsize=20)
plt.tight_layout()

plt.show()
```

### 比较两个班的成绩分布的特点
<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;两个班成绩都稍微右偏态，但是乙班看起来更严重，乙班成绩分布得更均匀一些
</font>

### 画出雷达图，看看两个班考试成绩的分布是否相似
<font color=#0099ff>
&ensp;&ensp;&ensp;&ensp;两个班成绩分布不相似
</font>

```python
# 
angles = np.linspace(0, 2*np.pi, 5, endpoint=False)
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, polar=True)
plt.plot(list(angles)+[0], data_jia['人数'].tolist() + data_jia['人数'].tolist()[0:1], '-og')
plt.plot(list(angles)+[0], data_yi['人数'].tolist() + data_yi['人数'].tolist()[0:1], '-*r')
plt.legend(['甲班', '乙班'])
_ = plt.thetagrids(angles*180/np.pi,data_jia['成绩'].tolist())
```
